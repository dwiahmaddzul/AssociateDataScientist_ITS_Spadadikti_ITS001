# Scatter Plot

Read Iris data

```python
import pandas as pd

irispath = r'drive/My Drive/Mengajar/Analitika Bisnis/Code/Dataset/2-5 iris-UCI-header.csv'
iris = pd.read_csv(irispath)

print(iris.columns.tolist())
```

Plot two class of Iris Data based on Petal Width & Petal Length Variable

List of colormap: https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
import matplotlib.pyplot as plt

#Setting up the scatter plot

print(iris)

iris.loc[iris['Species']=='Iris-virginica','Species']=0
iris.loc[iris['Species']=='Iris-versicolor','Species']=1
iris.loc[iris['Species']=='Iris-setosa','Species'] = 2

print(iris)

iris_2c = iris[iris['Species']!=2]
X = iris_2c['PetalLength'].values.T    # values.T transpose the columns into rows
Y = iris_2c['PetalWidth'].values.T

print(iris)

color = iris_2c[['Species']].values.T
color = color.astype('uint8')

# Draw the scatter plot

plt.scatter(X, Y, c=color[0,:], s=40, cmap=plt.cm.Spectral);
plt.title("IRIS DATA | Blue - Versicolor, Red - Virginica ")
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```
