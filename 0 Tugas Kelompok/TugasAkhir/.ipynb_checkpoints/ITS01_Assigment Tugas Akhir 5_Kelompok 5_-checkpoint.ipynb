```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn 
```

```python
Data ='Import-ekspor-banten.csv'
df =  pd.read_csv(Data,decimal=',')
```

```python
df.head()
```

```python
df.plot()
plt.show()
```

```python
cdf = df[['Nilai_Eksport','Nilai_Import']]
cdf.head(10)
```

```python
# Mengambil "Nilai Eksport" sebagai variabel independen (regressor)
X = cdf.iloc[:, 0].values
# Mengambil "Nilai Import" sebagai variabel dependen
y = cdf.iloc[:, 1].values
# Reshape data karena hanya menggunakan satu fitur
X = X.reshape(-1,1)
# Reshape data karena hanya satu fitur
y = y.reshape(-1,1)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
# Fitting SVR to the dataset
from sklearn.svm import SVR
regressor = SVR(kernel = 'linear')
#linear = Linear Kernel
regressor.fit(X, y)
```

```python
# Predicting a new result

y_pred = regressor.predict(sc_X.transform(np.array([[5.4]])))
#To transform 5.4 to the scaled X value, we first need to convert it into the array form
#Since the transform method of StandardScaler Library only accepts arrays

y_pred = sc_y.inverse_transform(y_pred)
#Now the prediction gives us the scaled value of y
#Thus we need inverse transformation of the scaled value for the real results
#nonlinear
#polinomial
#ordenya jangan tinggi max 4
print(y_pred)
```

```python
# Visualising the SVR results
plt.scatter(X, y, color = 'blue')
plt.plot(X, regressor.predict(X), color = 'red')
plt.title('Dataset Eksport-Import')
plt.xlabel('Eksport')
plt.ylabel('Import')
plt.show()
```

```python
# Visualising the SVR results
plt.scatter(sc_X.inverse_transform(X), sc_y.inverse_transform(y), color = 'blue')
plt.plot(sc_X.inverse_transform(X), sc_y.inverse_transform(regressor.predict(X)), color = 'red')
plt.title('Dataset Eksport-Import')
plt.xlabel('Eksport')
plt.ylabel('Import')
plt.show()
```

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(cdf[['Nilai_Eksport']])
test_y = np.asanyarray(cdf[['Nilai_Import']])
test_y_ = sc_y.inverse_transform(regressor.predict(sc_X.transform(test_x)))

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_ , test_y) )
```
